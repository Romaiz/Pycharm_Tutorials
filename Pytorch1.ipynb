```python
#Basic Networks
# Let xi be the independent variable height of each person in a group of i people, and let yi be the dependant variable weight
x = torch.tensor([[6,2], [5, 9], [5,4], [5,8], [5,1]]).float()
y = torch.tensor([72, 59, 61, 63, 54]).float()
x
```

```python
y
```

```python
M1 = nn.Linear(2,10, bias= False)
M1
```

```python
M2 = nn.Linear(10,1, bias= False)
M2
```

```python
M2(M1(x)).squeeze()
```

```python
#The above prediction model barely matches what we require, hence, we will apply a neural network to it 
class MyNeuralNetwork(nn.Module):
  def __init__(self): #This initalizes the class itself, i.e. contructor
    super().__init__() #Initializes everything that is in nn.Module
    self.Matrix1 = nn.Linear(2,10,bias=False) #Matrix1 and Matrix2 are defined as ATTRIBUTES of the class
    self.Matrix2 = nn.Linear(10,1,bias=False)
  def forward(self, x):
    x = self.Matrix1(x)
    x = self.Matrix2(x)
    return x.squeeze() 
```

```python
f = MyNeuralNetwork()
y_pred = f(x)
y_pred
```

```python
L = nn.MSELoss()
L(y_pred, y)
```

```python
#As we can see out error is still very big
optimizer = SGD(f.parameters(), lr= 0.001) #lr = learning rate
```

```python
losses = []
for _ in range(100):
  optimizer.zero_grad() # makes sure that gradient only accumaltes from a certain point and doesn't keep building up forever
  loss_value = L(f(x), y) # compute loss
  loss_value.backward() # compute gradient
  optimizer.step() # Adjust all the parameters
  losses.append(loss_value.item())
```

```python
plt.plot(losses)
plt.xlabel('epochs')
plt.ylabel('Loss')
```
